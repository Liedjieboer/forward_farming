```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/nico_marais/Downloads/USC00046074.csv")
```

```python
df.tail()
```

```python
df_napa = df[["NAME","DATE","TAVG","TMAX","TMIN"]]
```

```python
df_napa.info()
```

```python
df_napa["DATE"] = pd.to_datetime(df_napa["DATE"])
```

```python
df_napa.info()
```

```python
df_filtered = df_napa[df_napa["DATE"] >= "1978-01-01"]
```

```python
df_filtered
```

```python
df_filtered["day"] = df_filtered["DATE"].map(lambda x: x.day)
df_filtered["month"] = df_filtered["DATE"].map(lambda x: x.month)
df_filtered["year"] = df_filtered["DATE"].map(lambda x: x.year)
```

```python
df_filtered.drop(columns="DATE")
```

```python
df_df = df_filtered.groupby("year").mean().drop(columns=["day","month"])
df_df
```

```python
napa_score = ['9','9','7','8','9','9','8','8','7','8','9','9','10','7','8','8','8','8','9','7','8','7','10','6','8','8','6','8','9','9','7','5','8','7','10','4','4','6','5','6','6']
napa_score.reverse()
```

```python
df_df = df_df.head(41)
df_df["score"] = np.array(napa_score)

df_df["score"] = pd.to_numeric(df_df["score"])
```

```python
df_df.head()
```

```python
sns.regplot(x=df_df["TMIN"], y=df_df["score"])
```

```python
sns.regplot(x=df_df["TMAX"], y=df_df["score"], color="orange")
```

```python
sns.regplot(x=df_df["TAVG"], y=df_df["score"], color="green")
```
