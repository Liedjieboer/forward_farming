```python
import numpy as np
import pandas as pd
# read the image stack
data = pd.read_csv('better_sheff_merge.csv')
```

```python
x = data['x'].to_list()
y = data['y'].to_list()
z = data['z'].to_list()
```

```python
len(x), len(y)
```

```python
from mpl_toolkits.mplot3d.axes3d import *
import matplotlib.pyplot as plt
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter3D(x,y,z,c=z)  
plt.show()
```
