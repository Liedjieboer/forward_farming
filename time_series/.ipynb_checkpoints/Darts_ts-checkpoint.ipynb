```python
import pandas as pd
import numpy as np
from darts import TimeSeries
import calendar
from darts.models import ExponentialSmoothing
```

```python
data = pd.read_csv('../clean_data/clean_sheff_data.csv')
```

```python
data.tail(10)
```

```python
data['mm'] = data['mm'].apply(lambda x: calendar.month_abbr[x])
```

```python
data['yyyy'] = data['yyyy'].astype(str)
```

```python
data['period'] = data[['yyyy', 'mm']].agg('-'.join, axis=1)
```

```python
data
```

```python
series = TimeSeries.from_dataframe(data, "period", "avgtemp(degC)")
train, val = series[:-36], series[-36:]
model = ExponentialSmoothing()
model.fit(train)
prediction = model.predict(len(val), num_samples=1000)
```

```python
import matplotlib.pyplot as plt
```

```python
series.plot()
prediction.plot(label="forecast", low_quantile=0.05, high_quantile=0.95)
plt.legend()
```

```python
prediction1 = model.predict(1000, num_samples=1000)
series.plot()
prediction1.plot(label="forecast", low_quantile=0.05, high_quantile=0.95)
plt.legend()
```
