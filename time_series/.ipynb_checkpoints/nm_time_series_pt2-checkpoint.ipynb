```python
import pandas as pd
from darts import TimeSeries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read a pandas DataFrame
path = "/Users/nico_marais/code/Liedjieboer/forward_farming/clean_data/clean_sheff_data.csv"
df = pd.read_csv(path, delimiter=",")
```

```python
df.head()
```

```python
# Create a TimeSeries, specifying the time and value columns
series = TimeSeries.from_dataframe(df, "Unnamed: 0", "tmin(degC)")

# Set aside the last 36 months as a validation series
train, val = series[:-36], series[-36:]
```

```python
from darts.models import NaiveSeasonal

seasonal_model = NaiveSeasonal(K=12)
seasonal_model.fit(train)
seasonal_forecast = seasonal_model.predict(36)

series.plot(label="actual")
seasonal_forecast.plot(label="naive forecast (K=12)")
plt.xlim(left=1600)
```

```python
from darts.metrics import mae, mape
```

```python
from darts.models import ExponentialSmoothing, TBATS, AutoARIMA, Theta


def eval_model(model):
    model.fit(train)
    forecast = model.predict(len(val))
    print("model {} obtains MAE: {:.2f}".format(model, mae(val, forecast)))
    print("model {} obtains MAPE: {:.2f}%".format(model, mape(val, forecast)))


eval_model(ExponentialSmoothing())
```

```python
from darts.models import LinearRegressionModel

eval_model(LinearRegressionModel(lags=36))
```

```python
# from darts.models import Prophet

# eval_model(Prophet())
```

```python
from darts.models import RandomForest

eval_model(RandomForest(lags=12))
```

```python
from darts.models import RegressionModel

eval_model(RegressionModel(lags=36))
```

```python
from darts.models import RNNModel

eval_model(RNNModel(input_chunk_length=12))
```

```python
from darts.models import StatsForecastAutoARIMA

eval_model(StatsForecastAutoARIMA())
```

```python
from darts.models import StatsForecastETS

eval_model(StatsForecastETS())
```
